**Fuente de información:**

Usaremos el dataset de **INSURANCE COMPANY BENCHMARK (COIL 2000)**, el cual está disponible en:
<br/>

https://archive.ics.uci.edu/ml/datasets/Insurance+Company+Benchmark+%28COIL+2000%29
